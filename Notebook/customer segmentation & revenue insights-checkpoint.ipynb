# Feature Engineering(RFM )

    - Recency: Days since last purchase
    -Frequency: Number of unique invoices
    -Monetary: Total spending

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# last date in dataset

last_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
# Features table

RFM = df.groupby('Customer ID').agg({'InvoiceDate': lambda x: (last_date- x.max()).days,
                                     'Invoice': 'nunique',
                                     'TotalPrice': 'sum'
    
}    
).reset_index()
```

```python
# Renaming columns

RFM.columns = ["CustomerID", "Recency", "Frequency", "Monetary"]
```

# RFM Scoring & Segmentation

```python
# Scoring for recency, frequency and monetary

# Spliting into 5 parts and labeling(1,2,3,4,5) for frequency and monetary while for recency(5,4,3,2,1) most recent gets the highest score

RFM['r_score'] = pd.qcut(RFM['Recency'], 5, labels=[5,4,3,2,1], duplicates='drop')

RFM['f_score'] = pd.qcut(RFM['Frequency'], 5, labels=[1,2,3,4,5], duplicates='drop')

RFM['m_score'] = pd.qcut(RFM['Monetary'], 5, labels=[1,2,3,4,5], duplicates='drop')


# Skewed not enabling split
```

```python

RFM['r_score'] = pd.qcut(RFM['Recency'].rank(method='first'), 5, labels=[5,4,3,2,1])



RFM['f_score'] = pd.qcut(RFM['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])


RFM['m_score'] = pd.qcut(RFM['Monetary'].rank(method='first'), 5, labels=[1,2,3,4,5])
```

```python
# Converting to intergers

RFM['r_score'] = RFM['r_score'].astype(int)

RFM['f_score'] = RFM['f_score'].astype(int)

RFM['m_score'] = RFM['m_score'].astype(int)
```

```python
# Creating a combine rfm_score but concatenating the 3 scores as string


RFM['rfm_score'] = RFM['r_score'].astype(str) + RFM['f_score'].astype(str) + RFM['m_score'].astype(str)
```

```python
RFM.head()
```

```python
RFM.to_excel('RFM_complete_table.xlsx', index=False)
```

```python
# Creating the segment column

# Creating the parameters for each segment

def RFM_segment(row):
    if 	row['r_score'] >=4 and row['f_score'] >=4 and row['m_score'] >=4:
        return 'VIP customer'
    elif 	row['r_score'] >=4 and row['f_score'] >=4:
        return 'High value cuustomer'
    elif 	row['f_score'] >=4:
        return 'Loyal cuustomer'
    elif 	row['r_score'] <=2:
        return 'Chrun risk cuustomer'
    else:
        return 'Regular customer'

```

```python
# Creating the segment columns

RFM['Segment'] = RFM.apply(RFM_segment, axis = 1)

# Checking for counts in each segment

RFM['Segment'].value_counts()
```

# The largest segment (32.44%) consists of churn risk customers, highlighting the need for targeted campaigns to re-engage and win them back. Meanwhile, VIP and loyal customers (over 35%) should receive focused attention to maintain their loyalty and maximize long-term value.

# Segment KPIs KPIs by segment

    -Number of customers: Number of customers
    -Total revenue : Total revenue
    -Avg order value: Average spend per purchase
    -Total frequency: Total purchases
    -Avg recency: Average days since last purchase

```python
# Creating the KPIs

Segment_KPI = RFM.groupby('Segment').agg({ 'CustomerID': 'count', 
                                      'Monetary': 'sum',        
                                      'Frequency': 'sum',      
                                      'Recency': 'mean'         
                                     })\
                        .rename(columns={'CustomerID': 'Numbers of Customers',
                                                        'Monetary': 'Total Revenue',
                                                        'Frequency': 'Total purchases',
                                                        'Recency': 'Avg Recency'
                                                       })\
                        .round(2)\
                        .sort_values(by= 'Avg Recency')

Segment_KPI['Avg Order Value'] = (Segment_KPI['Total Revenue']/Segment_KPI['Total purchases']).round(2)


Segment_KPI
```

```python
Segment_KPI.to_excel('Segment_KPI.xlsx', index=False)
```

# Visual Insights
